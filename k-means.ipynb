```python
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.metrics import silhouette_score
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.metrics import silhouette_score
```

```python
# 读取四张表的数据
prior = pd.read_csv("D:/study/ml/数据/instacart/order_products__prior.csv")
```

```python
products = pd.read_csv("D:/study/ml/数据/instacart/products.csv")
```

```python
orders = pd.read_csv("D:/study/ml/数据/instacart/orders.csv")
```

```python
aisles = pd.read_csv("D:/study/ml/数据/instacart/aisles.csv")
```

```python
# 合并四张表到一张表  （用户-物品类别）
_mg = pd.merge(prior, products, on=['product_id', 'product_id'])
_mg = pd.merge(_mg, orders, on=['order_id', 'order_id'])
mt = pd.merge(_mg, aisles, on=['aisle_id', 'aisle_id'])
```

```python
mt.head(10)
```

```python
# 交叉表（特殊的分组工具）
cross = pd.crosstab(mt['user_id'], mt['aisle'])
```

```python
cross.head(10)
```

```python
# 进行主成分分析
pca = PCA(n_components=0.9)
```

```python
data = pca.fit_transform(cross)
data.shape
```

```python
# 把样本数量减少
# x = data[:500]
x = data
x.shape
```

```python
# 假设用户一共分为4个类别
km = KMeans(n_clusters=4)
km.fit(x)
```

```python
predict = km.predict(x)
predict
```

```python
# 显示聚类的结果
plt.figure(figsize=(10,10))
```

```python
# 建立四个颜色的列表
color = ["orange","green","blue","purple"]
# 遍历
col = [color[i] for i in predict]
plt.scatter(x[:,1],x[:,20],color=col)
plt.show()
```

```python
# 评判聚类效果，轮廓系数
silhouette_score(x,predict)
```
