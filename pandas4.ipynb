```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from matplotlib import font_manager
```

```python
file_path="D:/study/ml/数据分析资料/数据分析资料/day05/code/books.csv"
df=pd.read_csv(file_path)
print(df.head(1))
print(df.info())
```

```python
#求不同年份书的数量
#求不同年份书的平均评分
```

```python
data1=df[pd.notnull(df["original_publication_year"])]
grouped=data1.groupby(by="original_publication_year").count()["id"]
print(grouped)
_x=grouped.index
_y=grouped.values
plt.figure(figsize=(14,9),dpi=80)
plt.bar(range(len(_x)),_y)
plt.xticks(range(len(_x)),_x)
plt.show()
```

```python
data2=df[pd.notnull(df["original_publication_year"])]
grouped2=data2.groupby(by="original_publication_year")["average_rating"].mean()
print(grouped2)
_x=grouped2.index
_y=grouped2.values
plt.figure(figsize=(14,9),dpi=80)
plt.plot(range(len(_x)),_y)
plt.xticks(range(len(_x)),_x,rotation=270)
plt.show()
```
