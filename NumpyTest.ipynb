```python
import numpy as np
from matplotlib import pyplot as plt

us_filepath="D:/study/ml/数据分析资料/数据分析资料/day04/code/youtube_video_data/US_video_data_numbers.csv"
gb_filepath="D:/study/ml/数据分析资料/数据分析资料/day04/code/youtube_video_data/GB_video_data_numbers.csv"

us_data=np.loadtxt(us_filepath,delimiter=",",dtype="int")
gb_data=np.loadtxt(gb_filepath,delimiter=",",dtype="int")
```

```python
#取评论的数据
us_data_comment=us_data[:,-1]
print(us_data_comment.max(),us_data_comment.min(),)

d=5000
bin_nums=(us_data_comment.max()-us_data_comment.min())//d
us_data_comment=us_data_comment[us_data_comment<=5000]

#绘图
plt.figure(figsize=(14,9),dpi=80)

plt.hist(us_data_comment,bin_nums)
plt.show()
```

```python
gb_data=gb_data[gb_data[:,1]<=500000]
gb_data_comment=gb_data[:,-1]
gb_data_like=gb_data[:,1]
plt.figure(figsize=(14,9),dpi=80)
plt.scatter(gb_data_like,gb_data_comment)
plt.show()
```
